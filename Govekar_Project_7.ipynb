Support Vector Machines or SVM are,in machine learning, "supervised learning models with associated learning algorithms that analyze data used for classification and regression analysis"[5]. Which means that SVMs assist in assigning new element entries based on prior data, as long as the given data has a pretty large sample set, and the different patterns and trends that we examined from it. We essentially want to make use of the machine learning methods that assist us in making forcasts of classifications for new data sets, based on the prior patterns and trends.

One advantage of using SVMs is their predtictability. The ability to use the data we possess to forecast patterns and trends makes SVMs very reliable in the field of computing and coding. The disadvantage is that the given data must be unsystematic, albeit  it's probable for a data set  to have an inadvertent bias which could turn out to be of no valuable use. For instance, constructing a data set that emphasizes too much on one specific area or has far too much of data for a single variable and very limited data for other variables could be dentrimental to the forecasting process and the predictions we want.

# 2.2

```python
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

irisModel = SVC()
```

```python
X = irisData.drop(['species'], axis=1)
y = irisData['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

We chose the 70-30 split above. So 70% of the data is selected in a random order for training and the remaining 30% of the data tests the model after training it. So we didn't introduce bias in the dataset. This process is called cross-validation which uses the built in function in scikit-learn 
What we will do here is the following. We will split up the data set. We will keep a lot of data for training the model. But we will keep some of it aside to test our model. We will need to be careful about how we do this. The sampling has to be done randomly. We cannot introduce bias into either dataset. This process is called cross-validation. We used a built-in function called test_train_split() in scikit-learn for this.

# 1.3

```python
irisModel.fit(X_train, y_train)
```

# 1.4

```python
predictions = irisModel.predict(X_test)
#printing the confusion matrix
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

When we analyze the above data, we should understand the meanings of 'precision' and "recall". Looking at the 3x3 confusion matrix, the diagonal element positions of (1,1), (2,2) and (3,3) are representative of the predicted numbers of the Iris setosa, Iris vesicolor and Iris virginicas. Any other numbers placed beside the diagonals are outliers and are either anomalies or species whos identification were incorrect.

Utilizing the same, we will show the confusion matrix via the classification report. The report offers us the percentages of the data and precision is defined by the total number of true positives divided by the total numer of true positives plus the total number of false positives

$$ precision = \frac{total\,true\,positives}{total\,true\,positives\,+\,total\,false\,positives}$$

Recall uses a similar equation, barring that the denominator is the total number of false negatives rather than false positives.

$$ recall = \frac{total\,true\,positives}{total\,true\,positives\,+\,total\,false\,negatives}$$

Whereas f1-score is calculated as follows.

$$f_1 \ score = 2 \ \frac{precision\times recall}{precision + recall}$$
