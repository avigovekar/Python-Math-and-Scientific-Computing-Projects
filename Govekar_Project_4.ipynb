# Introduction
In this project, we will study the behavior of a gas in a closed system by the Van der Waal's equation and its reduced forms.  Johannes Diderik van der Waals, a Dutch theoretical physicist and thermodynamicist, proposed the model for gases in 1857. The Van der Waal's gas equation has an advantage over the ideal gas equation since it incorporates two properties of an actual realistic gas i.e. gas molecules have size and gas molecules interact with each other. Following is the Van der Waal's equation:

$(P +\frac{an^2}{V^2})(V − b) = nRT$

where P is gas pressure, V is the volume of the container that encloses the gas and T is the temperature of the gas. The constants a and be are the values that represent interactions between molecules and their size respectively. The Van der Waal's equation specifically takes into account, the non ideal behaviour of gases i.e. real gases because in real world situations, and reduces the chances of assumptions and errors. Assumptions like "volume taken by imaginary ideal gas molecules" and "gas molecules don't attract or repel each other" can be ignored according to Van der Waal's theory.

We will use two different methods, Bisection method and Newton-Raphson method to find out the different roots of the Van der Waal's equation to understand phase change of gas and critical point of the gas.

```python
def p_R(v_R,T_R):
    """
    Function returns Van der Waal's gas equation (3) below
    
    Keywords:
        T_R : Temperature of gas
        v_R : Volume of the container that entraps the gas

    """
    
    return  ((8*T_R)/(3*v_R-1))-(3/(v_R*v_R))
```

```python
def f(v_R=0.1):
    """
    Function returns the output given below.  
    
    Keywords:
    Default v_R = 0.1
    
    """
    return ((-1.2*v_R**3)+(6.8*v_R**2)-(9*v_R)+3)
```

```python
def bisectionFunc(f, a, b, tolerance):
    """
    Returns the root of the input function f and the number of iterations it takes as
    a pair using the bisection method.
    
    Keywords:
    f: function for which root is to be found. Default is myFunction
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    
    """
    # First figure out signs of a and b
    signa = np.sign(f(a))
    signb = np.sign(f(b))
    iter = 0
    
    # Write loop for the bisection method. Don't stop till tolerance is met
    while b-a > tolerance:
        c = (a+b)/2
        print("Best guess of root = {} at iteration {:3d}".format(c, iter))
        
        signc = np.sign(f(c))
        
        # What happens if c is zero? It means root is found!
        if signc == 0:
            return c
        elif signc == signa:
            a = c
        else:
            b = c
            
        iter += 1
    
    # Return final value
    return c, iter
```

```python
#1st root
bisectionFunc(f, 0.1, 1, tolerance=1e-7)
```

```python
#2nd root
bisectionFunc(f, 1, 2, tolerance =1e-7)
```

```python
#3rd root
bisectionFunc(f, 3, 4, tolerance=1e-7)
```

```python
def fx(v_R=0.1):
    """
    Function returns the output given below and its derivative as a pair of values.  
    
    Keywords:
    Default v_R = 0.1
    
    """
    
    # Return function value and derivative as a pair
    return ((-1.2*v_R**3)+(6.8*v_R**2)-(9*v_R)+3),((-3.6*v_R**2)+(13.6*v_R)-9)
```

```python
def newtonMethod(fx, xInit , tolerance):
    """
    Returns the root of the input function fx and the number of iterations it takes as
    a pair using the Newton's method.
    
    Keywords:
    fx: function for which root is to be found. Default is myFunction2. fx must
       return both function value and derivative at x.
    a: left most point of bracket [a, b]. Default is 0.4
    b: right most point of bracket [a, b]. Default is 1.0
    tolerance: the decimal point up to which precision is desired. Default is 1e-7
    
    """
    
    # Initialize the variables
    step = 0
    f, df = fx(xInit)
    dx = -f/df
    x = xInit
    
    # Run the iterations in the algorithm
    while np.abs(dx) > tolerance:        
        print("Best guess of root = {} at iteration {:3d}".format(x, step))
        f, df = fx(x)
        dx = -f/df
        x += dx
        step += 1
    
    # Return output
    return x, step
```

```python
newtonMethod(fx, xInit = 0, tolerance = 1e-7)
```

```python
newtonMethod(fx, xInit = 1, tolerance = 1e-7)
```
